```python
import spacy
nlp = spacy.load('en')

def tokenize(line):
    return [(token.text, token.tag_) for token in nlp(line)]


def extract_ingredients_nouns(line):
    ingredients_nouns = set()
    token_tag_pairs = tokenize(line)
    for pair in token_tag_pairs:
        if pair[1] == 'NN' or pair[1] == 'NNS':
            # global ingredients_nouns
            ingredients_nouns |= {pair[0]}

    return list(ingredients_nouns)
```

```python
import re
import requests
from bs4 import BeautifulSoup

url = 'https://www.allrecipes.com/recipe/8665/braised-balsamic-chicken/'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')

# extract ingredients section from the webpage
lines = set([element.label.text.strip() for element in soup.find_all(class_='checkList__line')])
# remove unnecessary elements
lines.remove('')
lines.remove('Add all ingredients to list')
lines

def get_ingredients(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')

    # extract ingredients section from the webpage
    lines = set([element.label.text.strip() for element in soup.find_all(class_='checkList__line')])
    # remove unnecessary elements
    lines.remove('')
    lines.remove('Add all ingredients to list')
    lines
    return lines
```

```python
import random

def get_random(ingres):
    l = len(ingres)
    if l == 0 or (l == 1 and ingres[0] is None):
        return ''
    ran_num = random.randint(0,l-1)
    res = ingres[ran_num]
    
    # handle None type
    if res is None:
        return get_random(ingres)
    else:
        return res

print(get_random(list(vegie_proteins.keys())))
```

```python
# some special ingredients
special_sub = {
    'frankfurter':'vegetarian sausage',
  'meatball':'vegetarian meatball',
  'meatloaf':'tofu',
  'sausage':'vegetarian sausage',
  'bacon':'vegetarian bacon',
  'ham':'vegetarian ham',
  'skewer':'tofu skewer'}

def get_vegie_substitution(ingre):
    sub = ''
    if 'stock' in ingre or 'consomme' in ingre or 'bouillon' in ingre:
        sub = 'vegetable stock'
    elif 'broth' in ingre:
        sub = 'vegetable broth'
    else:           
        for meat in list(special_sub.keys()):
            if meat in ingre or ingre in meat:
                sub = special_sub[meat]
    if sub == '':
        # choose a random vegie protein
        sub = get_random(vegie_protein_list)
    return sub
    
def get_meat_substitution(ingre):
    if is_broth(ingre):
        return 'chicken broth'
    for meat in list(special_sub.keys()):
        if meat in ingre or ingre in meat:
            return meat
    return get_random(meat_protein_list)
```

```python
def is_broth(ingre):
     if 'stock' in ingre or 'consomme' in ingre or 'bouillon' in ingre or  'broth' in ingre:
        return True
     return False

def is_meat(ingre):
    for meat in meat_protein_list:
        if meat in ingre or ingre in meat:
            return True
    if is_broth(ingre) is True:
        return True
    else:
        return False
    return False

def is_vegie_protein(ingre):
    for i in vegie_protein_list:
        if i in ingre or ingre in i:
            return True
    if is_broth(ingre) is True:
        return True
#     if 'stock' in ingre or 'consomme' in ingre or 'bouillon' in ingre or  'broth' in ingre:
#         return True
    return False
```

# To Vegetarian 

```python
res = []
url = 'https://www.allrecipes.com/recipe/219763/chicken-marsala-over-white-rice/'
# lines = get_ingredients(url)
recipe = utils.Recipe(url)
```

```python
# go through the directions first to get a substitution dict
directions = recipe.directions
sub_dict = {}

for direction in directions:
#     print(direction)
    single_direction_ingredients = recipe.extract_ingredients(direction)
    for ingre in single_direction_ingredients:
        if is_meat(ingre) is True and ingre not in sub_dict.keys():
            sub = get_vegie_substitution(ingre)
            sub_dict[ingre] = sub
print(sub_dict)
    
```

```python

for line in recipe.ingredients:
    quantity, measurement, ingredient_name, preparation = recipe.extract_all(line)
    descriptor = recipe.extract_descriptor(ingredient_name)
    
    # remove descriptor if not None
    ingredient = ingredient_name.replace(descriptor, '').strip() if descriptor else ingredient_name
    # if ingredient is empty after removing descriptor
    if ingredient == '':
        ingredient = ingredient_name
    # remove ' to taste' in ingredient if any
    ingredient = re.sub(r'(or)? to taste', '', ingredient)
    ingredient = ' '.join(ingredient.split())
        
    if is_meat(ingredient) is True:
        sub = ''
        for key in list(sub_dict.keys()):
            if key in ingredient or ingredient in key:
                sub = sub_dict[key]
        if sub == '':
            sub = get_vegie_substitution(ingredient)
            sub_dict[ingredient] = sub
            
        if is_broth(ingredient) is True:
            sub_measurement = measurement
            sub_preparation = '' if preparation else ''
            sub_descriptor = '' if descriptor else ''
            
        else:
            sub_measurement = get_random(list(vegie_proteins[sub]['measurement']))
            sub_preparation = get_random(list(vegie_proteins[sub]['preparation']))
            sub_descriptor = get_random(list(vegie_proteins[sub]['descriptor']))
            
        replacement = quantity+' '+sub_measurement+' '+sub_descriptor+' '+sub
        print('REPLACE')
        print(line)
        print('TO')
        print(replacement)
    else:
        print(line)
    print('----------------------------')
        

print(sub_dict)
```

```python
for direction in directions:
    new_direction = direction
    for key in list(sub_dict.keys()):
            if key in direction:
                new_direction = new_direction.replace(key, sub_dict[key])
                
    # go over the meat ingredient in case words like 'breast' / 'chop' /
    for word in new_direction.split():
        for meat in meat_protein_list:
            if word == meat or meat+'s' == word:
                new_direction = new_direction.replace(word, '')
                
    print('-----------------------------------------------------------')
    if direction != new_direction:
        print('REPLACE')
        print(direction)
        print('TO')
        print(new_direction)
    else:
        print(direction)
```
