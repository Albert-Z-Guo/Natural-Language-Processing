```python
import json
import pprint

# 3 pages of vegetarian protein recipe ingredients
# 1 pages of vegetarian recipe ingredients
# 1 pages of all 7 meat and poultry categries
# 2 pages of seafood recipe ingredients

with open('vegeratian_ingredient.json', 'r') as f:
    ingredients = json.load(f)
    pprint.pprint(ingredients)
```

```python
# names with description
vege_ingredient = ingredients['vegetarian']
seafood_ingredient = ingredients['seafood']
meat_ingredient = ingredients['meat_and_poultry']
vege_protein_ingredient = ingredients['vegetarian_protein']

# noun
seasonings = ingredients['seasonings']
seasonings = list(set(seasonings))
```

```python
import spacy
nlp = spacy.load('en')

def tokenize(line):
    return [(token.text, token.tag_) for token in nlp(line)]


def extract_ingredients_nouns(line):
    ingredients_nouns = set()
    token_tag_pairs = tokenize(line)
    for pair in token_tag_pairs:
        if pair[1] == 'NN' or pair[1] == 'NNS':
            # global ingredients_nouns
            ingredients_nouns |= {pair[0]}

    return list(ingredients_nouns)
```

```python
vege_protein_dict = {}
for i in vege_protein_ingredient:
    l = extract_ingredients_nouns(i)
    for ingre in l:
        # eliminate seasonings
        if ingre in seasonings or ingre == '':
            continue
            
        if ingre in vege_protein_dict:
            vege_protein_dict[ingre] += 1
        else:
            vege_protein_dict[ingre] = 1
# print(vege_protein)
sorted_d =  sorted(vege_protein_dict.items(), key=lambda d:d[1], reverse = True)
        
# print(vege_protein_dict)
print(sorted_d)
```

```python
meat_ingre_dict = {}
for i in meat_ingredient + seafood_ingredient:
    l = extract_ingredients_nouns(i)
    for ingre in l:
        # eliminate seasonings
        if ingre in seasonings or ingre == '':
            continue
            
        if ingre in meat_ingre_dict:
            meat_ingre_dict[ingre] += 1
        else:
            meat_ingre_dict[ingre] = 1
# print(vege_protein)
sorted_d =  sorted(meat_ingre_dict.items(), key=lambda d:d[1], reverse = True)
        
# print(vege_protein_dict)
print(sorted_d)
```

```python
vege_protein_ingredient = list(vege_protein_dict.keys())
meat_ingredient = list(meat_ingre_dict.keys())
```

```python
common_ingredients = set(vege_protein_ingredient) & set(meat_ingredient)
print(common_ingredients)
```
