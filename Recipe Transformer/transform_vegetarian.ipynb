```python
# load selected meat list
meat_dict = {}
with open('selected_meat_ingredients.txt', 'r') as file:
    for line in file:
        meat_dict[line.strip()] = {}
file.close()

# populate ingredients' measurements, descriptors, and preparations from scraped data
# we combined both meat and seafood
for i in meat_dict:
    if i in ingredients['meat']['measurement']:
        meat_dict[i]['measurement'] = ingredients['meat']['measurement'][i]
    if i in ingredients['meat']['descriptor']:
        meat_dict[i]['descriptor'] = ingredients['meat']['descriptor'][i]
    if i in ingredients['meat']['preparation']:
        meat_dict[i]['preparation'] = ingredients['meat']['preparation'][i]

for i in meat_dict:
    if i in ingredients['seafood']['measurement']:
        meat_dict[i]['measurement'] = ingredients['seafood']['measurement'][i]
    if i in ingredients['seafood']['descriptor']:
        meat_dict[i]['descriptor'] = ingredients['seafood']['descriptor'][i]
    if i in ingredients['seafood']['preparation']:
        meat_dict[i]['preparation'] = ingredients['seafood']['preparation'][i]
        
# for ingredients not in scraped data, manually add general data
for i in meat_dict:
    if i not in ingredients['meat']['measurement']:
        meat_dict[i]['measurement'] = ['pound']
    if i not in ingredients['meat']['descriptor']:
        meat_dict[i]['descriptor'] = ['fresh', 'raw', None]
    if i not in ingredients['meat']['preparation']:
        meat_dict[i]['preparation'] = ['cut into cubes']
        
pprint.pprint(meat_dict)
```

```python
# load selected vegetarian protein list
vegetarian_dict = {}
with open('selected_vegetarian_ingredients.txt', 'r') as file:
    for line in file:
        vegetarian_dict[line.strip()] = {}
file.close()

for i in vegetarian_dict:
    if i in ingredients['vegetarian_protein']['measurement']:
        vegetarian_dict[i]['measurement'] = ingredients['vegetarian_protein']['measurement'][i]
    else:
        vegetarian_dict[i]['measurement'] = ['cup', 'can']
    if i in ingredients['vegetarian_protein']['descriptor']:
        vegetarian_dict[i]['descriptor'] = ingredients['vegetarian_protein']['descriptor'][i]
    else:
        vegetarian_dict[i]['descriptor'] = [None]
    if i in ingredients['vegetarian_protein']['preparation']:
        vegetarian_dict[i]['preparation'] = ingredients['vegetarian_protein']['preparation'][i]
    else:
        vegetarian_dict[i]['preparation'] = ['drained']

pprint.pprint(vegetarian_dict)
```

```python
def randomly_select(ls):
    if len(ls) == 0 or (len(ls) == 1 and ls[0] is None):
        return ''
    choice = random.choice(ls)
    if choice is None:
        return randomly_select(ls)
    else:
        return choice

# test
print(randomly_select(list(vegetarian_dict.keys())))
```

```python
# custom ingredients substitutions
custom_to_vegetarian_dict = {'frankfurter':'vegetarian sausage',
                            'meatball':'vegetarian meatball',
                            'meatloaf':'tofu',
                            'sausage':'vegetarian sausage',
                            'bacon':'vegetarian bacon',
                            'ham':'vegetarian ham',
                            'skewer':'tofu skewer'}

custom_from_vegetarian_dict = {v: k for k, v in custom_to_vegetarian_dict.items()}

def get_vegetarian_substitution(ingredient):
    if 'stock' in ingredient or 'consomme' in ingredient or 'bouillon' in ingredient:
        return 'vegetable stock'
    if 'broth' in ingredient:
        return 'vegetable broth'
    if ingredient in custom_to_vegetarian_dict:
        return custom_to_vegetarian_dict[ingredient]
    # randomly substitue a vegetarian protein if no custom substitution is found
    return randomly_select(list(vegetarian_dict.keys()))
        
def get_meat_substitution(ingredient):
    if is_broth(ingredient):
        return 'chicken broth'
    if ingredient in custom_from_vegetarian_dict:
        return custom_from_vegetarian_dict[ingredient]
    return randomly_select(list(meat.keys()))
```

```python
def is_broth(ingredient):
     if 'stock' in ingredient or 'consomme' in ingredient or 'bouillon' in ingredient or  'broth' in ingredient:
        return True
     return False

def is_meat(ingredient):
    if ingredient in meat_dict:
        return True
    if is_broth(ingredient):
        return True
    return False

def is_vegetable(ingredient):
    if ingredient in vegetarian_dict:
        return True
    if is_broth(ingredient):
        return True
    return False
```

# Transform to Vegetarian 

```python
url = 'https://www.allrecipes.com/recipe/219763/chicken-marsala-over-white-rice/'
recipe = utils.Recipe(url)
```

```python
sub_dict = {}

for line in recipe.ingredients:
    quantity, measurement, descriptor, ingredient, preparation = recipe.extract_all(line)
        
    # if ingredient is meat
    if is_meat(ingredient):
        sub = get_vegetarian_substitution(ingredient)
        sub_dict[ingredient] = {}
        
        # if ingredient is broth
        if is_broth(ingredient):
            sub_measurement = measurement
            sub_preparation = None
            sub_descriptor = None
        else:
            sub_measurement = randomly_select(list(vegetarian_dict[sub]['measurement']))
            sub_descriptor = randomly_select(list(vegetarian_dict[sub]['descriptor']))
            sub_preparation = randomly_select(list(vegetarian_dict[sub]['preparation']))
        
        # record substitution information
        sub_dict[ingredient]['substitution'] = sub
        sub_dict[ingredient]['measurement'] = sub_measurement
        sub_dict[ingredient]['descriptor'] = sub_descriptor
        sub_dict[ingredient]['preparation'] = sub_preparation
        
        # pick not None element only
        replacement = []
        for i in [quantity, sub_measurement, sub_descriptor, sub]:
            if i is not None:
                replacement.append(i)        
        
        print('\nREPLACE:')
        print('\t' + line)
        print('TO:')
        print('\t' + ' '.join(replacement) + '\n')
    else:
        print(line)
```

```python
# expand sub_dict() for better granularity
sub_dict_granular = {}
for ingredient in sub_dict.keys():
    if len(ingredient.split()) > 1: 
        for word in ingredient.split():
            sub_dict_granular[word] = {}
            sub_dict_granular[word]['substitution'] = sub_dict[ingredient]['substitution']
            sub_dict_granular[word]['measurement'] = sub_dict[ingredient]['measurement']
            sub_dict_granular[word]['descriptor'] = sub_dict[ingredient]['descriptor']
            sub_dict_granular[word]['preparation'] = sub_dict[ingredient]['preparation']
```

```python
pprint.pprint(sub_dict)
pprint.pprint(sub_dict_granular)
```

```python
replaced = set()
for direction in recipe.directions:
    new_direction = direction
    
    bulk_substitution = False
    for ingredient in sub_dict.keys():
        if ingredient in direction:
            new_direction = direction.replace(ingredient, sub_dict[ingredient]['substitution'])
            bulk_substitution = True
            for i in sub_dict[ingredient]['substitution'].split():
                replaced |= {i}
            print("Replaced '{0}' to '{1}':".format(ingredient, sub_dict[ingredient]['substitution']))

    if not bulk_substitution:
        for word in sub_dict_granular.keys():
            if word in new_direction and word not in replaced:
                new_direction = new_direction.replace(word, sub_dict_granular[word]['substitution'])
                for i in sub_dict_granular[word]['substitution'].split():
                    replaced |= {i}
                print("Replaced '{0}' to '{1}':".format(word, sub_dict_granular[word]['substitution']))
                
    if direction != new_direction:
        print('REPLACE:')
        print('\t' + direction)
        print('TO:')
        print('\t' + new_direction)
    else:
        print(direction)
    print('-----------------------------------------------------------\n')
```
